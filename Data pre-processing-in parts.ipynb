```python
import numpy as np
import pandas as pd
```

```python
dataFrame = pd.read_csv("labeledTrainData.tsv", header=0, delimiter="\t", quoting=3)
dataFrame.head()
```

```python
dataFrame.review[0]
```

```python
# Removing HTML syntaxes
from bs4 import BeautifulSoup as bs
```

```python
data = bs(dataFrame.review[0], "lxml")
data.get_text()
```

```python
# removing punctuation marks
import re
# Use regular expressions to do a find-and-replace
letters_only = re.sub("[^a-zA-Z]",           # The pattern to search for
                      " ",                   # The pattern to replace it with
                      data.get_text() )  # The text to search
print(letters_only)
```

```python
# convert to lower case
lowerCase = letters_only.lower()
lowerCase
```

```python
# splitting into words
words = lowerCase.split()
len(words)
# words
```

```python
# dir(letters_only)
```

```python
# Removing stop-words
import nltk
from nltk.corpus import stopwords
```

```python
print(len(stopwords.words("english")))
sw = stopwords.words("english")
```

```python
words = [w for w in words if w not in sw ]
len(words)
# words
```
